## Regression Analysis

- predicting conyinous valued outcome by using provided input variables(predictors, independent variables)
- we need to establish a relationship between dependent (output/target variables) with the independent variable
- in regression we have 2 types:
1. Linear Regression: 
- here basic assumption is data is linear (both dependent and independent variable has linear relationship
- the linear relationship can be positive or negative

2. Non linear Regression


- y = mx+c, m is coefficient and c is intercept
- in order to find out coefficient we can use various method like ordinary list square and gradient descient algoritm, neural networks

1. model MLalgorithm (y = mx + c)
2. set initial values of the coefficients
3. set the cost function (to find errors)
    - for regression it is MSE(mean squared error)
    - Cf = 1/n * sum(actual-predicted) ^ 2
4. find out the fradient using partial derivatives on CF
5. repeat the step 2 to 5 until covergenc (till we reach minimum error) or till we reach the max iteration


```python
import matplotlib.pyplot as plt

Enginesize = [2.0, 2.4, 1.5, 3.5, 3.5, 3.5, 3.5, 3.7, 3.7]
c02emission = [196, 221, 136, 255, 244, 230, 232, 255, 267]
predicted = [203, 218.6, 183.5, 261.5, 261.5, 261.5, 261.5, 269.3, 269.3]

plt.scatter(Enginesize, c02emission, marker = '*', color = 'red', label = 'actual data line')


plt.plot(Enginesize, predicted, marker = '*', label = 'regression line')
plt.legend()
plt.show()
```
