# Neural Network

### input layer: 
   - it accecpt the user input along with initial weight in the model
   - inputs will be multiplied by weights summed together and passed to the next layer
    
### hidden layer:
   - it is the complex structure cabable to complex hidden pattern inside the data
   - no specific rule to select it
   - half of the size of input neural size

### output layer:
   - here we get the final output process based on the input process of previous layer

### activation function: 
   - it is resposible for introducing non lenearity forr the model without activation neural network become linear model
   - mathematically activation function decide wether a neuron wil be active or not by calculating the weighted sum and adding bias to it
   

## Steps

- input passes from input layer via activation reaches hidden layer and then finally output layer with learnt output (based on initial weight of model)
- compute error between model predicted value as actual value
- Backward propagation: using the error computed backward propagate the error so as to updated the weights of model
- process of forward and bakward propagation is called one iteration
- Iteration: one time forward and backward propagation of single sample in the data
- Epoch: one time forward and backward propagation of entire data is called one Epoch
- Batch size: no of samples in each iteration
- ReLu will be most preferred in hiddden layer

```python
import keras
from keras.models import Sequential  # we will build sequential architecture
from keras.layers import Dense       # for adding the layers to sequential model
```

```python
import numpy as np
Ain  = np.array([1,2,3,4,5,6])
Bout = np.array([4,5,6,7,8,9])
print(Ain)
print(Bout)
```

```python
Reg_model = Sequential()
Reg_model.add(Dense( units = 1, input_shape = [1]))
Reg_model.compile(optimizer = 'sgd', loss = 'mean_squared_error' )
```

```python
Reg_model.summary()
```

```python
Reg_model.fit(Ain, Bout, epochs=500)

# try for 1 epoch, check result
# try for 10 epoch. check result
# 100 check result
# 500 check result
```
