# Visualization with matplotlib

## 1.line plot
- pass the x axis values and y axis values
- additional pass information like color, line width, line style etc
- further we can add label for axises like x,y and titles
- you may also try controlling the angle, rotation etc

```python
a = [1,2,3,4,5,6,7,8]
b = [10,20,30,40,50,60,70,80]

# make sure length of x and y axis are same
print(len(a), len(b))
```

```python
plt.plot(a,b, marker = '*')
plt.show()  #to display figure
```

```python
import numpy as np
# sin(2*pi*f*t)
# f-frequency t-time period

pi = 3.14
f = 1
t = np.arange(0,1,0.05)
y = np.sin(2*pi*f*t)
#print(t,y)
plt.plot(t,y, color = 'red', linewidth = 1, marker = '*', 
         mfc = 'blue', linestyle = 'dashed', ms = 10, mec = 'y') 
#mfc-marker face color, mec-marker edge color, ms-marker size

plt.xlabel('Time period')
plt.ylabel('Amplitude')
plt.title('Sinewave')
plt.show()
```

```python
np.random.seed(0)
a = np.arange(1,11,1)
b = np.random.randint(1,10,10)
c = np.random.rand(10)
plt.plot(a,c,'cd-.')  # x axis is linear 
# r red, * marker style, - line style
plt.show()

plt.plot(b,c,'m+--') #x axis is not linear
plt.show()

# when x axis is not in particular order do not use line plot instead 
# an also when dat is large 

plt.scatter(b,c)
plt.show()
```

## 2.Bar plot
- useful when we have x axis data in the string format y axis in number
- comparing different countries population
- x axis country name
- y axis population value

- comparing scores of students
- x : student names
- y : scores

```python
countries = ['India', 'Bangladesh', 'china', 'USA', 'Brazil']
population = [145,22,140,35,26]

plt.title('Countrywise population') # is used when more than one visulization in single plot
plt.bar(x = countries, height = population)
plt.show()

plt.title('Countrywise population')
plt.bar(x = countries, height = population, color = 'orange', edgecolor = 'green', width = .3)
plt.xticks(rotation = 45)
plt.show()
```

```python
countries = ['India', 'Bangladesh', 'china', 'USA', 'Brazil']
population = [145,22,140,35,26]
GDP = [3338,460,1790,2500,192]


plt.title('Countrywise population')
plt.bar(x = countries, height = GDP, color = 'cyan')
plt.bar(x = countries, height = population, color = 'orange', edgecolor = 'green', width = .6)
plt.yticks([10])
plt.xticks(rotation = 45)
plt.show()

# secound bar chart will over write first one,
# as values in secound are larger than first one
# vertical stacked plots, horizontal stacking 

#--------------vertical stack---------------------
plt.bar(x = countries, height = GDP, color = 'cyan')
plt.bar(x = countries, height = population, bottom = GDP, color = 'orange', width = .6)
plt.show()

#---------------horizontal stack--------------------
xindex = np.arange(len(countries))
plt.bar(x = xindex, height = GDP, color = 'cyan', width = 0.3)
plt.bar(x = xindex+0.3, height = population, color = 'orange', width = .3)
plt.xticks(xindex+0.3/2, countries)
plt.show()
```
