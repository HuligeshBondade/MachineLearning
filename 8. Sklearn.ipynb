```python
# before splitting the data perrform outlier treatment first

df_num = df.select_dtypes(include = np.number)
df_cat = df.select_dtypes(exclude = np.number)
```

```python
# Missing data handling imputation on to numerical coiumn

from sklearn.impute import SimpleImputer
Imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
df_num.iloc[:,:] = Imputer.fit_transform(df_num.iloc[:,:])
df_num
```

```python
# categorical data handling

df_cat1 = pd.get_dummies(df_cat, drop_first = True)
df_cat1
```

```python
# Standard scaler for numerical column

df_num
```

```python
# standard scaler = (x - mean of column) / sd of column
# min-max scaler = (x - xmin) / (xmax - xmin)
standard scaler : maintains the original distribution of data even after scaling
min-max scaler : it keeps the data in the range of 0 to 1 scale 
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
df_num.iloc[:,:] = SS.fit_transform(df_num)
df_num
```

```python
# now colab the data together

df_cleaned = pd.concat([df_num, df_cat1], axis = 1)
df_cleaned
```
