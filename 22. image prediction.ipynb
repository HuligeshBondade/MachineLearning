```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
```

```python
model = ResNet50(weights='imagenet')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
img1 = r'C:\Users\Huligesh\Desktop\Abeyaantrix Internship\Dataset\dog\img3.jpg'
img_path = img1
img = image.load_img(img_path, target_size = (224,224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
preds = model.predict(x)

# showing the image with prediction
plt.figure(figsize=(15,20))
img_bgr = cv2.imread(img_path, 1)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.title('Prediction using RestNet50 pre-training model:{}'.format(decode_predictions(preds,top=1)[0]))
```
