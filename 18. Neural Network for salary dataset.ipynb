```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

path = r'C:\Users\Huligesh\Desktop\Abeyaantrix Internship\Dataset'
file = r'\Salary_Data.csv'
df = pd.read_csv(path + file)
df.head()
```

```python
x = df.iloc[:, 0:1]
y = df.iloc[:, 1:]
```

```python
import keras
from keras.models import Sequential  # we will build sequential architecture
from keras.layers import Dense       # for adding the layers to sequential model
```

```python
Reg_model = Sequential()
Reg_model.add(Dense( units = 1, input_shape = [1]))
Reg_model.compile(optimizer = 'sgd', loss = 'mean_squared_error' )
```

```python
Reg_model.fit(x,y,epochs = 1000)
```

```python
Reg_model.predict(x.iloc[0])
```
