```python
[R,C] = df.shape
print('Dataset has %d rows and %d column'%(R,C))
print('below is the list of column\n',df.columns)
```

```python
# what are highest and lowest levels of co2 emission 
co2max = df.CO2EMISSIONS.max()
co2min = df.CO2EMISSIONS.min()

print('''Heighest and lowest of co2 emission in the given dataset are %d and %d'''%(co2max,co2min))
```

```python
# how many car brands are present and what are the names of the same 

print('Car brand names = \n',df.MAKE.unique())
print('there are total %d brands are present' %(len(df.MAKE.unique())))
```

```python
# types of fuel used and which is uesd more
print(df.FUELTYPE.unique())
print(df.FUELTYPE.value_counts())
print(df.FUELTYPE.max())
print('-'*25)

# average of engine size
print(df.ENGINESIZE.mean())
print('-'*25)

# how many car engine size more than 4
es = df.ENGINESIZE > 4
print(df[['ENGINESIZE']][es])
print('car engines are',df.ENGINESIZE[df.ENGINESIZE > 4].count())
print('-'*25)

#how many cars have less than 3 cylinders
print('cylinders have',df.CYLINDERS[df.CYLINDERS < 3].count())
```

```python
# engine size vs co2 emission

plt.figure(figsize = (4,4))
plt.plot(df.ENGINESIZE, df.CO2EMISSIONS)
plt.show()
```

```python
plt.figure(figsize = (4,4))
plt.scatter(df.ENGINESIZE, df.CO2EMISSIONS)
plt.show()
print('As the engine size increases co2 also increases')
```

```python
# cylinders vs co2 emission scatter plot

plt.figure(figsize = (4,4))
plt.scatter(df.CYLINDERS, df.CO2EMISSIONS)
plt.show()
print('As the cylinders increases co2 also increasing')
```

```python
# fueltype vs co2 emission

plt.figure(figsize = (4,4))
plt.bar(df.FUELTYPE, df.CO2EMISSIONS)
plt.show()
```

```python
# make vs co2 emission

plt.figure(figsize = (10,4))
plt.bar(df.MAKE, df.CO2EMISSIONS)
plt.xticks(rotation = 90)
plt.show()
```
